```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from bokeh.plotting import figure, show, output_file
from bokeh.palettes import brewer
import bokeh
```

```python
df = pd.read_csv('/home/andrew/Documents/ML/Cape_Town_water_levels/year_dam_data.csv', index_col='Year')
df
```

```python
berg_river = 130010
steenbras_lower = 33517
steenbras_upper = 31767
theewaterskloof = 480188
voëlvlei = 164095
wemmershoek = 58644
dams_volume = [berg_river]
df = df.applymap(lambda x: round(x/6, 2))
df.head()
```

```python
def stacked(df):
    df_top = df.cumsum(axis=1)
    df_bottom = df_top.shift(axis=1).fillna({df.columns[0]: 0})[::-1]
    df_stack = pd.concat([df_bottom, df_top], ignore_index=True)
    return df_stack
areas = stacked(df)
```

```python
colors = brewer['PuBu'][areas.shape[1]]
x2 = np.hstack((df.index[::-1], df.index))

p = figure(title="Cape Town water levels", x_range=(2008, 2018), y_range=(0, 100))
p.grid.minor_grid_line_color = '#eeeeee'
p.ygrid.minor_grid_line_alpha = 0.05

p.patches([x2] * areas.shape[1], [areas[c].values for c in areas],
          color=colors, alpha=0.8, line_color=None, legend='one')

p.plot_height = 400
p.plot_width = 1000

p.xaxis.axis_label = 'Year'
p.yaxis.axis_label = '% Water'
p.xaxis.minor_tick_line_color = None  # turn off x-axis minor ticks

output_file('water_vis.html', title='Cape_Town')
show(p)
```
