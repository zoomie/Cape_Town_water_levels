```python
import pandas as pd
import numpy as np

from bokeh.io import output_notebook, show, output_file
from bokeh.plotting import figure
from bokeh.palettes import Spectral11

# output_file('new.html', title='hey')
output_notebook()
```

```python
df = pd.read_csv('/home/andrew/Documents/ML/Cape_Town_water_levels/year_dam_data.csv', index_col='Year')
# df['Year'] = df['Year'].apply(lambda x: x*)
# df = df.set_index('Year')
df.head()
```

```python
# toy_df = pd.DataFrame(data=np.random.rand(5,3), columns = ('a', 'b' ,'c'), index = pd.DatetimeIndex(start='01-01-2015',periods=5, freq='d'))   
numlines=len(df.columns)
mypalette=Spectral11[0:numlines]
```

```python
p = figure(width=500, height=300, x_axis_type="datetime") 
p.xaxis.axis_label = 'Year'
p.yaxis.axis_label = '% Water'
p.multi_line(xs=[df.index.values]*numlines,
                ys=[df[name].values for name in df],
                line_color=mypalette,
                line_width=5)
show(p)
```
